```python
import pandas as pd
from sqlalchemy import create_engine

DATABASE_URL = "postgresql://postgres:12345678@localhost:5432/medical_warehouse"
engine = create_engine(DATABASE_URL)

files = [
    "E:/Shipping-a-Data-Product-From-Raw-Telegram-Data-to-an-Analytical-API-week_8/data/csv/telegram_messages/2026-01-17/CheMed123_utf8.csv",
    "E:/Shipping-a-Data-Product-From-Raw-Telegram-Data-to-an-Analytical-API-week_8/data/csv/telegram_messages/2026-01-17/doctorfasil_utf8.csv",
    "E:/Shipping-a-Data-Product-From-Raw-Telegram-Data-to-an-Analytical-API-week_8/data/csv/telegram_messages/2026-01-17/lobelia4cosmetics_utf8.csv",
    "E:/Shipping-a-Data-Product-From-Raw-Telegram-Data-to-an-Analytical-API-week_8/data/csv/telegram_messages/2026-01-17/tikvahpharma_utf8.csv"
]

for file in files:
    df = pd.read_csv(file, encoding='utf-8')

    # Rename columns to match PostgreSQL table
    df.rename(columns={
        'date': 'message_date',
        'text': 'message_text',
        'views': 'view_count',
        'forwards': 'forward_count'
    }, inplace=True)

    # Optional: ensure types match (timestamp, int, boolean, etc.)
    df['message_date'] = pd.to_datetime(df['message_date'], errors='coerce')
    df['view_count'] = pd.to_numeric(df['view_count'], errors='coerce')
    df['forward_count'] = pd.to_numeric(df['forward_count'], errors='coerce')
    df['has_media'] = df['has_media'].astype(bool)

    # Insert into PostgreSQL
    df.to_sql('telegram_messages', engine, schema='raw', if_exists='append', index=False)

print("All CSVs loaded successfully!")
```
